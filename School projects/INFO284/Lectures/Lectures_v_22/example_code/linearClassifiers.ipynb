```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples = 1000, noise = 0.25, random_state = 3)
```

```python
import mglearn
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
log_reg = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 12)
import time

start = time.time()
log_reg.fit(X_train,y_train)
stop = time.time()
logreg_time = stop-start
```

```python
print("Training set score: {:.3f}".format(log_reg.score(X_train,y_train)))
print("Test set score: {:.3f}".format(log_reg.score(X_test,y_test)))
print("Time: {:.3f}".format(logreg_time))
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
mglearn.plots.plot_2d_separator(log_reg,X,eps = 0.5)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
start = time.time()
svc.fit(X_train,y_train)
stop = time.time()
svc_time = stop-start
print("Training set score: {:.3f}".format(svc.score(X_train,y_train)))
print("Test set score: {:.3f}".format(svc.score(X_test,y_test)))
print("Time: {:.3f}".format(svc_time))
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
mglearn.plots.plot_2d_separator(svc,X,eps = .5)
```

```python
from sklearn.svm import SVC
#svm = SVC(kernel = 'poly', degree=3, C = 1, gamma = 1)
svm = SVC(kernel = 'rbf', C = 10, gamma = 1)

start = time.time()
svm.fit(X_train,y_train)
stop = time.time()
svm_time = stop-start
print("Training set score: {:.3f}".format(svm.score(X_train,y_train)))
print("Test set score: {:.3f}".format(svm.score(X_test,y_test)))
print("Time: {:.3f}".format(svm_time))
```

```python
import mglearn
mglearn.plots.plot_2d_separator(svm,X,eps = .5)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```
