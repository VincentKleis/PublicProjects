```python
import networkx as nx
G = nx.karate_club_graph()
```

task 1.1

```python
nodes = list(G.nodes.items())

print(nodes)
```

```python
edges = list(G.edges.items())

print(edges)
```

task 1.2

```python
clubs = []
for i in nodes:
    index, attrs = i
    if attrs['club'] not in clubs:
        clubs.append(attrs['club'])

print(clubs)
```

```python
colors = [0 if node[1]['club'] == 'Mr. Hi' else 1 for node in nodes]

options = {
    "node_color" : colors,
    "cmap" : "spring"
}
```

```python
nx.draw_networkx(G, **options)
```

task 1.3

```python
path = nx.dijkstra_path(G, 26, 16)

print(path)
```

task 1.4

```python
colors = []
for node in nodes:
    if node[0] in path:
        colors.append(0)
    elif node[1]['club'] == 'Mr. Hi':
        colors.append(1)
    else:
        colors.append(2)

options = {
    "node_color" : colors,
    "cmap" : "autumn"
}
for node in nodes:
    if node[0] in path:
        G.nodes[node[0]]['color'] = 'red'
```

```python
nx.draw_networkx(G, **options)
```
