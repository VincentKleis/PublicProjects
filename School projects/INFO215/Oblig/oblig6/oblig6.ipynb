```python
from networkx import Graph, draw_networkx

G = Graph()

with open('nutrients.csv') as file:
    nutrients = file.readlines()

for x in nutrients:
    x = x.rstrip('\n').split(',')
    G.add_edge(x[0], x[1])

draw_networkx(G)
```

```python
from networkx import degree_centrality, modularity_matrix
centrality = degree_centrality(G)
# modularity using Louvain method
modularity = modularity_matrix(G)

modularity = [sum(x) for x in modularity]

modularity = sorted(modularity)
print(modularity)
```
